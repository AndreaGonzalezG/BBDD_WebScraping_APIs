Abrir archivo json en pandas

```python
import json
```

```python
with open("personal.json", "r") as fichero: # introducimos el nombre de nuestro archivo json y "r" porque lo vamos a leer(read)
    personal = json.load(fichero)
```

```python
print(json.dumps(personal,indent=4))
```

```python
personal["workers"]
```

```python
personal_pandas={'nombre':[], 'edad':[], 'experiencia':[], 'nivel': [], 'salario':[]}

for persona in personal["workers"]:
   
   personal_pandas['nombre'].append(persona['nombre'])
   personal_pandas['edad'].append(persona['edad'])
   personal_pandas['experiencia'].append(persona['experiencia'])
   personal_pandas['nivel'].append(persona['nivel'])
   personal_pandas['salario'].append(persona['salario'])


print(personal_pandas)
```

```python
# Actualizar el diccionario 'personal' con 'personal_pandas'
personal.update(personal_pandas)
```

```python
with open("personal.json", "w") as json_file:
    json.dump(personal, json_file)
```
